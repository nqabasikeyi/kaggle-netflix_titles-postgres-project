```python
from pathlib import Path
import pandas as pd

def load_netflix_data(csv_path=Path(r"C:\Users\AMAZONCA1\Desktop\git-proj-kaggle\kaggle-netflix_titles-postgres-project\data\raw\netflix_titles.csv")):
    csv_path = Path(csv_path)

    if not csv_path.is_file():
        raise FileNotFoundError(
            f"Can't find CSV at: {csv_path.resolve()}\n"
            f"Current working directory: {Path.cwd()}"
        )

    return pd.read_csv(csv_path)

df = load_netflix_data()
display(df.head())
```

```python
# Display the Datset structure
print("\nDataset structure:")
print(df.info())
```

```python
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
print("\nColumn Names:")
print(df.columns)
```

```python
print("\nDataset Shape:")
print(df.shape)
```

```python
# Display the number of unique values in each column
print("\nUnique Values per Column:")
print(df.nunique())
```
